# A Quick Exploration of SPY/VIX Relationships

As traders and investors, any new information has the potential to provide an edge in the market. While it is commonplace to use past price data to predict market direction, implied volatility can give fresh insights into the current market environment and help improve our decisions. This notebook will take a quick look at how the S&P index price is affected by its volatility, or VIX. In particular, we'll look at how VIX is related to recent historical volatility and actual realized volatility.

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

First we'll upload a csv of historical prices for both the CBOE Volatility Index (VIX) and S&P 500 ETF (SPY). Prices used are the closing prices for both indices, adjusted for splits and dividends. The 'Rets' column contains the percentage returns received if one were to buy on the next day's open and sell on the following day's open, as follows:

$$Return_t = \frac{O_{t+2}}{O_{t+1}} - 1$$

where O represents the symbol's opening price. We'll use all the data Yahoo! Finance has available, which extends back to January, 1993.

```python
df = pd.read_csv('VIX Study.csv', parse_dates=True, index_col=0).sort_index(ascending=True).round(2)
df['SPY returns'] = df['SPY Close'].pct_change()
df['VIX returns'] = df['VIX'].pct_change()
df['fwd returns'] = df['SPY Open'].pct_change().shift(-2)
df['hist vol'] = df['SPY returns'].rolling(21).std() * np.sqrt(252) * 100
df['fwd vol'] = df['SPY returns'].rolling(21).std().shift(-21) * np.sqrt(252) * 100
df.head(8)
```

We can look at the correlations between the three measures of volatility to see how each affect one another. A couple of interesting facts here. First, historical volatility and implied volatility are highly correlated. If one knew nothing else, the best estimate for volatility during the next 21 days is what it was the past 21 days. Second, implied volatility appears to be a better predictor of realized volatility than historical volatility. Useful information for traders trying to determine the likely trading range over the next month or so.

```python
volatilities = df[['VIX', 'hist vol', 'fwd vol']].dropna()
volatilities.corr()
```

```python
sns.pairplot(volatilities)
```

```python
df[['SPY returns', 'VIX returns']].corr()
```

The first relationship we'll examine is the spread, or difference, between the VIX and the 21-day forward-looking realized volatility. If this spread is consistently positive, it indicates that investors expected volatility to be higher than it actually was. Effectively, they were willing to pay more for insurance than "fair value" according to the [Black-Scholes model](https://en.wikipedia.org/wiki/Black%E2%80%93Scholes_model). If the spread is negative, the market turned out to be more risky than expected.

```python
vix_fwd = volatilities['VIX'] - volatilities['fwd vol']
vix_fwd.hist(bins=20)
plt.title('Implied Volatility - Realized Volatility')
plt.xlabel('% Difference')
plt.ylabel('Occurences')
```

```python
vix_fwd.describe()
```

As can be seen in the histogram above, the spread was largely positive during our examination period. Most of the time, the options traded at a premium. However, there some nasty fat tails at the left edge of the distribution. These were times where investors significantly underestimated the volatility in the market, characteristic of [Black Swan Events](http://blackswanevents.org/?page_id=26). We can highlight the days on which volatility was the most under-priced to see what the market environment typically looked like. No surprises here: it appears that the days where volatility was underestimated most severely occurred during the bursting of the Dot Com Bubble and the Global Financial Crisis.

```python
bottom_percentile = np.percentile(vix_fwd, 2.5)
worst_days = vix_fwd < bottom_percentile
df.dropna()['SPY Close'].plot()
df.dropna().loc[worst_days, 'SPY Close'].plot(style='ro')
```

Next up, we'll take a look at the spread between historical volatility and implied volatility. As noted earlier, these two are highly correlated, so any large deviations may be noteworthy. A positive spread means investors think the near future will more volatile than the recent past, and vice versa.

```python
vix_hist = volatilities['VIX'] - volatilities['hist vol']
vix_hist.hist(bins=20)
```

Similar to our past findings, we see that implied volatility is typically higher than historical volatility. And those pesky fat tails showed up again! Markets are most interesting at the extremes, so let's see what those market environments look like.

```python
vix_hist_low = np.percentile(vix_hist, 1)
vix_hist_high = np.percentile(vix_hist, 99)
vix_hist_low_days = vix_hist <= vix_hist_low
vix_hist_high_days = vix_hist > vix_hist_high
df.dropna()['SPY Close'].plot()
df.dropna().loc[vix_hist_low_days, 'SPY Close'].plot(style='ro')
df.dropna().loc[vix_hist_high_days, 'SPY Close'].plot(style='go')
```

This chart's slightly more unclear, although it does appear that these days tend to cluster around market bottoms. 

Finally, lets' take a look at how well VIX predicts the price range in the SPY going forward. The VIX gives an estimate for the forward annualized standard deviation. Assuming a normal distribution (which we've shown may not be a great idea), price should fall within 1 standard deviation approx. 68% of the time. If accurate, this range is represented by the following:

$$ C_{SPY, \, t+n} = C_{SPY, \, t} \times ( \, 1 \pm \frac{C_{VIX, t}}{100} \times \sqrt{n} \div \sqrt{252} ) \, $$

We'll examine how these predictions held up in volatile (2008) and quieter (2017) times.

```python
df['proj upper'] = df['SPY Close'].shift(21) * (1 + df['VIX'].shift(21) / 100 * np.sqrt(21) / np.sqrt(252))
df['proj lower'] = df['SPY Close'].shift(21) * (1 - df['VIX'].shift(21) / 100 * np.sqrt(21) / np.sqrt(252))
df.loc['2008', ['SPY Close', 'proj upper', 'proj lower']].plot(style=['b-', 'g:', 'r:'])
```

```python
df.loc['2017', ['SPY Close', 'proj upper', 'proj lower']].plot(style=['b-', 'g:', 'r:'])
```

We can also look at how many multiples of VIX away the price closed in the future. If the VIX accurately represents the future standard deviation and the errors are normally distributed, we should see most of these values all within $\pm 3 \: \times$ VIX.

```python
n = 5
expected_num = df['SPY Close'] - df['SPY Close'].shift(n)
expected_denom = df['SPY Close'].shift(n) * df['VIX'].shift(n) / 100 * np.sqrt(n) / np.sqrt(252)
df['vs expected'] = expected_num / expected_denom
df['vs expected'].hist()
```

```python
def sharpe_ratio(returns):
    return returns.mean() / returns.std()

def drawdown(eq):
    return (eq / eq.cummax() - 1)

print(sharpe_ratio(df['fwd returns']) * np.sqrt(252))
print(sharpe_ratio(df.loc[df['vs expected'] < -0.5, 'fwd returns']) * np.sqrt(252))
```

```python
drawdown(eq).plot()
df['SPY DD'] = drawdown(df['SPY Close'])
df['SPY DD'].plot()
```

```python
strat_ulcer = np.sqrt(((eq / eq.cummax() - 1)**2).mean()) * 100
spy_ulcer = np.sqrt(((df['SPY Close'] / df['SPY Close'].cummax() - 1)**2).mean()) * 100
print(strat_ulcer)
print(spy_ulcer)
```

```python
def summary_stats(returns):
    stats = pd.Series()
    gains = returns[returns > 0]
    losses = returns[returns <= 0]
    num_total = len(returns)
    num_gains = len(gains)
    num_losses = len(losses)
    avg = np.mean(returns)
    volatility = np.std(returns)
    sharpe = avg / volatility
    win_pct = num_gains / num_total
    avg_win = np.mean(gains)
    avg_loss = np.mean(losses)
    stats['total returns'] = num_total
    stats['total gains'] = num_gains
    stats['total losses'] = num_losses
    stats['expectancy (%)'] = avg * 100
    stats['volatilty (%)'] = volatility * 100
    stats['sharpe (daily)'] = sharpe
    stats['win %'] = win_pct * 100
    stats['total returns'] = num_total    
    stats['average gain (%)'] = avg_win * 100    
    stats['average loss (%)'] = avg_loss * 100
    return stats
    
```

```python
strat = df.loc[df['VIX'] < df['hist vol'], 'fwd returns']
pd.DataFrame({'strat':summary_stats(strat), 'SPY':summary_stats(df['fwd returns'])})
```

```python
import statsmodels.api as sm
X = df.dropna()['hist vol']
X = sm.add_constant(X)
y = df.dropna()['VIX']
model = sm.OLS(y, X).fit()
model.params
```

```python
model.summary()
```

```python
historical_component = df['hist vol'] * model.params[1] + model.params[0]
plt.scatter(df['hist vol'], df['VIX'])
plt.plot(df['hist vol'], historical_component, color='r')
```

```python
# plot residuals
resid = df['VIX'] - historical_component
resid.plot()
#plt.axhline(resid.mean(), color='r')
plt.title('VIX Residuals')
```

```python
def obj_func(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

def normalize(x):
    return (x - np.mean(x)) / np.std(x)

def bound(x):
    return (x - np.min(x)) / (np.max(x) - np.min(x))
```

```python
normalized = normalize(resid)
bounded = bound(normalized)
bounded = bounded * 2 - 1
bounded.plot()
plt.title('Scaled Residuals')
```

```python
print(len(bounded))
print(len(df['fwd returns']))
```

```python
df['fwd returns'].groupby(pd.qcut(bounded, 10)).mean().plot(kind='bar')
plt.title('Mean Return by Excess VIX Decile')
plt.xlabel('Scaled Excess VIX')
plt.ylabel('Mean Daily SPY Return')
```

```python
compare_df = pd.DataFrame()
compare_df['SPY'] = summary_stats(df['fwd returns'])
compare_df['Top Decile'] = summary_stats(df.loc[bounded > bounded.quantile(0.9), 'fwd returns'])
compare_df
```

```python
expanding_quantile = bounded.expanding(min_periods=10).quantile(0.9)
top_quantile = bounded > expanding_quantile
filtered = df.loc[top_quantile, 'fwd returns']
filtered_2 = df.loc[df['VIX']>30, 'fwd returns']
filtered_3 = df.loc[bounded > 0, 'fwd returns']
results = pd.DataFrame()
results['SPY'] = summary_stats(df['fwd returns'])
results['Top Quantile'] = summary_stats(filtered)
results['Excess > 0'] = summary_stats(filtered_3)
results['High VIX'] = summary_stats(filtered_2)
(1 + filtered).cumprod().plot()
(1 + filtered_2).cumprod().plot()
(1 + filtered_3).cumprod().plot()
(1 + df['fwd returns']).cumprod().plot()
results
```

```python
num_days = (bounded.index[-1] - bounded.index[0]).days
num_years = num_days / 365.25
total_return = (1 + 2 * filtered).cumprod().dropna()
cagr = (total_return.iloc[-1] / total_return.iloc[0]) ** (1 / num_years) - 1
max_dd = drawdown(total_return).min()
print('CAGR: {}'.format(cagr))
print('Max DD: {}'.format(max_dd))
```
